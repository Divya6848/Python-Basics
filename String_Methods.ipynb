```python
str = "Python"
print(type(str))

#String Operations
#Concatenation
sector = "B.tech"
stream = "Computer Science"
result = sector + " " + stream
print(result)

#Repitition
stream = "Mechanical "
result = stream * 22
print(result)

#Indexing
str = "Accessing individual characters in a string"
print(str[19])
print(str[-5])

#Slicing
str = "Extracting a Substring"
sub = str[7:20]
sub1 = str[3:]
print(sub)
print(sub1)
```

```python
#Python Format Strings
#OLD style formatting
name = "Divya"
age = 21
formatted_string = "My name is %s and I am %d yrs old."%(name,age)
print(formatted_string)

#Format Method
name = "Divya"
age = 21
formatted_string = "My name is {} and I am {} yrs old.".format(name, age)
print(formatted_string)

#String interpolation(f-string)
name = "Divya"
age = 21
formatted_string = f"My name is {name} and I am {age} yrs old."
print(formatted_string)
```

```python
#String Methods
#1.len()
language = "Python is fun"
length = len(language) #returns the length of the string
print(length)

#2.Lower()/Upper()
stream = "datascience is Good"
print(stream.upper()) #convert the string to lower/upper cases

stream = "DATASCIENCE IS GOOD"
print(stream.lower())

#3.Strip()
my_str = "     Jupyter      "
stripped_string = my_str.strip() #remove the leading and trailing whitespace
print(stripped_string)

#4.Replace()
stream = "Data, Science"
new_stream = stream.replace("Science", "Analyst") #replace a specified string
print(new_stream)

#5.Split()
movie = "Kalki movie is a Blockbuster Hit"
words = movie.split() #split the string at the specified separator
print(words)

#6.join()
movie = ["kalki", "movie", "is", "a", "Blockbuster", "Hit"]
join_words = ' '.join(movie)
print(join_words)
```

```python
#7.capitalize()  #capitalize the first character of the string
str1 = "prabhas in bahubali"
capital = str1.capitalize() 
print(capital)

#8.title()   #capitalize the first character of eqch word in the string
str2 = "prabhas in bahubali"
title = str2.title() 
print(title)

#9.lstrip()   #Returns a left trim version of the string
str3 = "   DATA SCIENCE   "
stripped_str = str3.lstrip() 
print(stripped_str)

#10.rstrip()   #Returns a right trim version of the string
str4 = "    DATA SCIENCE   "
stripped_str = str4.rstrip() 
print(stripped_str)
 
#11.swapcase()   #lower case becomes upper case and vice versa
game = "FooT bALl"
swap = game.swapcase()
print(swap)

#12.Casefold()   #Converts string into lower case
book = "Mahabaratha THE Epic"
case_str = book.casefold() 
print(case_str)

#13.center()   #Returns a centered string
sum = "BHAHUBALI"
centered_sum = sum.center(20)
print(centered_sum)

#14.count()   #the number of times a specified value occurs in a string
movies = "Prabhas in bhahubali"
count_movie = movies.count("b") 
print(count_movie)

str9 = "Astronauts are amazing!"
count_str = str9.count("a")
print(count_str)

#15.endswith()   #true if the string ends with the specified value
my_string = "UniverseGalaxy MilkywayGalaxy!"
print(my_string.endswith("Galaxy!")) 

```

```python
#16.expandstab()   #Sets the tab size of the string
text = "I\tN\tD\tI\tA"
expand_text =  text.expandtabs(4) 
print(expand_text)

#17.find()    #search the position of string ,where it was found
order = "zepto big basket"
find_order = order.find("basket") 
print(find_order)

# returns (-1) if string is not found
find_order = order.find("bsket") 
print(find_order)

#18.rfind()  # search the last position of string, where it was found
order = "zepto big basket mart"
find = order.rfind("mart")
print(find)

# (-1) if string not found
find = order.rfind("market")
print(find)

#19.index()   #search the position of string ,where it was found
orders = "zomato swiggy amazon mart"
found_order1 = orders.index("swiggy") 
print(found_order1)

#found_order2 = orders.index("basket") # returns "value error" if string is not found
#print(found_order2)

#20.rindex()    #search the last position of string ,where it was found
orderss = "zomato swiggy amazon mart"
founds_order = orderss.rindex("amazon") 
print(founds_order)

#21.isalnum()     #True if all char in the string are alphanumeric
sector = "DataScience1011"
num = sector.isalnum()
print(num)

sectors = "DataScience 10-11"
num = sectors.isalnum()
print(num)

#22.isalpha()    #True if all char in the string are in the alphabets
comment = "Adorablesweet"
alpha = comment.isalpha()
print(alpha)

comments = "adorable,sweet"
alpha = comments.isalpha()
print(alpha)

#23.isascii()   #True if all char in the string are ascii characters
brand = "1zara $gucci %max"
app = brand.isascii()
print(app)

#24.isdecimal()   #True if all char in the string are decimals
brands = "234556566"
apps = brands.isdecimal()
print(apps)

brands = "2345 56566"
apps = brands.isdecimal()
print(apps)

#25.isdigit()   #True if all char in the string are digits from various Unicode char
ph_num = "9988¹²33"
digit = ph_num.isdigit()
print(digit)
```

```python
#26.isidentifier()  #True if the string is a valid identifier
target = "special_character"
print(target.isidentifier()) 

target = "1special character"
print(target.isidentifier()) 

#27.islower() #True if all char are lower case
tour = "arya is a tourist"
print(tour.islower())

tour = "Arya is a Tourist"
print(tour.islower())

#28.isnumeric()  #true if all char are in numeric
dict = "75287633"
print(dict.isnumeric())

dict = "752, 87633"
print(dict.isnumeric())

#29.isprintable()
book = "fictional_123"
print(book.isprintable())

#30.isspace()  #True if all char in the string are whitespaces
stream = "   "
print(stream.isspace()) 

#31.istitle()   #True if the string follows the rules of a title
sum = "Statistics, Probability!"
print(sum.istitle())

#32.isupper() #true if all char in the string are upper case
tour = "AMERICAN TOURIST"
print(tour.isupper())

#33.ljust()  #Left-justifying a string
a = "cricket"
b = a.ljust(10)
print(b)

a = "cricket"
b = a.ljust(15, "$")
print(b)

#34.rjust() #right-justifying a string
p = "Avacado"
q = p.rjust(20)
print(q, "is my favorite fruit.")

p = "Avacado"
q = p.rjust(20, "@")
print(q, "is my favorite fruit.")

#35.maketrans()  #return a translation table to be used in translations
name = "Hello Divya!"
table = name.maketrans("D", "P")
print(name.translate(table))


```

```python
#36.partition()  #returns a tuple where the string is parted into three parts
sport = "I could play carrom all day"
z = sport.partition("carrom")
print(z)

#37.rpartition() #right partition
sport = "I could play carrom all day, carrom is my favorite game"
game = sport.rpartition("tennis")
print(game)

#38.rsplit() #right split
art = "carrom, tennis, football"
game = art.rsplit(",",1)
print(game)

#39.splitlines()   #Splits the string at line breaks
music = "I\nlove\nMusic!"
result = music.splitlines()
print(result)

#40.startswith()   #returns true if the string starts with the specified value
music = "Hii, Music is passion."
z = music.startswith("Hii")
print(z)

#41.translate()  #Returns a translated string
paint = "Hello Aoctor!"
mytable = str.maketrans("A", "D")
print(paint.translate(mytable))

#42.zfill()  #Fills the string with a specified number of 0 values at the beginning
num = "The moon is a symbol of hope"
sum = num.zfill(56)
print(sum)

```
